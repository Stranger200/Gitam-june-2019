```python
def createFile(name):
    f = open(name,"w")
    l = int(input("enter number of lines"))
    for i in range(l):
        e=input("enter %d line"%i)
        f.write("%s\n"%e)
    print("File is successfully created ")
    f.close()
    return

def readFile(name):
    f=open(name,"r")
    if f.mode=="r":
        x=f.read()
        print(x)
    f.close()
    return
def appendFile(name):
    f = open(name,"a")
    l = int(input("enter number of lines"))
    for i in range(l):
        e=input("enter %d line"%i)
        f.write("%s\n"%e)
    print("File is successfully created ")
    f.close()
    return
def wordCount(name,word):
    f=open(name,"r")
    if f.mode=="r":
        x=f.read()
        lst=x.split()
    cnt=lst.count(word)
    return cnt
def charCount(name):
    n=0
    f=open(name,"r")
    if f.mode=="r":
        x=f.read()
        lst=list(x)
        n=len(lst)
    return n-1 
def upperCount(name):
    n=0
    f=open(name,"r")
    if f.mode=="r":
        x=f.read()
        lst=list(x)
        for j in lst:
            if ord(j)>=65 and ord(j)<=90:
                n=n+1
    return n 
def lineCount(name):
    n=0
    f=open(name,"r")
    if f.mode=="r":
        x=f.read()
        lst=list(x)
        for j in lst:
            if ord(j)==10:
                n=n+1
    return n 

i=0
while(i!=1):
    c=int(input("enter choice\n1 for creating file file\n2 for veiwing the file\n3for adding to previous file\n4 for finding number of times a word is repeated\n5 for counting charecters\n6 for counting upper\n7 for counting number of lines\n8 for stopping\n"))
    if c==1:
        name=input("enter file name with extension")
        createFile(name)
    elif c==2:
        name=input("enter file name with extension")
        readFile(name)
    elif c==3:
        name=input("enter file name with extension")
        appendFile(name)
    elif c==4: 
        name=input("enter file name with extension")
        word=input("enter word to be counted")
        print("number of times given word is repeated is",wordCount(name,word))
    elif c==5:
        name=input("enter file name with extension")
        print("number of charecters is",charCount(name))
    elif c==6:
        name=input("enter file name with extension")
        print("number of upper case charecters is",upperCount(name))
    elif c==7:
        name=input("enter file name with extension")
        print("number of lines is",lineCount(name))
    else:
        i=1
        break
    
```

```python
#regular expressions
#[0-9]-->anydigits
#[a-z]-->small letter
#[A-Z]-->Capital letters
#[2468]-->all single digit even numbers
#^[0-9]{1}$-->only single digit numbers
#  .cap symbolis used for start of regular expression
#  .dollar symbol is used for regular expressions
#^[a-zA-Z]{5}$-->capitals or small with 5 charecters
#^[a-zA-Z]{6-15}$-->capitals or small with 6 to 15 charecters
#^[a].....[z]$-->any string of length 5 starts with a ends with z 
#^[a].*[z]-->any string of any length starts with a ends with z 
```

```python
import re
def phoneValidation(phone):
    pattern = "^[6-9][0-9]{9}$|^[0][6-9][0-9]{9}$|^[+][9][1][6-9][0-9]{9}$"
    phone = str(phone)
    if re.match(pattern,phone):
        return "valid"
    return "invalid"
phone=input("enter your mobile number")
print("it is a %s mobile number"%phoneValidation(phone))


```

```python
import re
def rollNumber(rn):
    rn=str(rn)
    pattern="^[1][5][2][U][1][A][0][1-9][0-6][0-9]$"
    if re.match(pattern,rn):
        return "valid"
    return "invalid"
rn=input("enter your mobile number")
print("it is a %s ROLL number"%rollNumber(rn))
```

```python
import re
def email(rn):
    rn=str(rn)
    pattern="^[0-9a-zA-Z][0-9a-zA-Z_.]{4,13}[0-9a-zA-Z][@][a-z0-9]{3,18}[.][a-z]{2,4}$"
    if re.match(pattern,rn):
        return "valid"
    return "invalid"
rn=input("enter your emailid ")
print("it is a %s emailid"%email(rn))
```

```python
import re
def password(rn):
    pattern="^[a-zA-Z0-9@$#!]{6,15}$"
    if re.match(pattern,rn):
        return "valid"
    return "invalid"
rn=input("enter your NEW PASSWORD")
print("it is a %s password"%password(rn))
```
