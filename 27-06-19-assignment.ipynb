```python
#assignment questions 1 to 5

def createFile(name):
    f = open(name,"w")
    l = int(input("enter number of lines"))
    for i in range(l):
        e=input("enter %d line"%i)
        f.write("%s\n"%e)
    print("File is successfully created ")
    f.close()
    return

def readFile(name):
    f=open(name,"r")
    if f.mode=="r":
        x=f.read()
        print(x)
    f.close()
    return
def appendFile(name):
    f = open(name,"a")
    l = int(input("enter number of lines"))
    for i in range(l):
        e=input("enter %d line"%i)
        f.write("%s\n"%e)
    print("File is successfully created ")
    f.close()
    return
def charCount(name):
    n=0
    f=open(name,"r")
    if f.mode=="r":
        x=f.read()
        lst=list(x)
        for j in lst:
            if j.isnumeric():
                n=n+1
    return n 
def upperCount(name):
    n=0
    f=open(name,"r")
    if f.mode=="r":
        x=f.read()
        lst=list(x)
        for j in lst:
            if j.islower():
                n=n+1
    return n 
def lineCount(name):
    n=0
    f=open(name,"r")
    if f.mode=="r":
        x=f.read()
        lst=list(x)
        for j in lst:
            if not j.isnumeric() and not j.isupper() and not j.islower() and not j.isspace() and ord(j)!=10:
                n=n+1
    return n 

i=0
while(i!=1):
    c=int(input("enter choice\n1 for creating file file\n2 for veiwing the file\n3for adding to previous file\n4 for finding number of digits\n5 for counting lower case charecters\n6 for counting special charecters\n8 for stopping\n"))
    if c==1:
        name=input("enter file name with extension")
        createFile(name)
    elif c==2:
        name=input("enter file name with extension")
        readFile(name)
    elif c==3:
        name=input("enter file name with extension")
        appendFile(name)
    elif c==4:
        name=input("enter file name with extension")
        print("number of digits is",charCount(name))
    elif c==5:
        name=input("enter file name with extension")
        print("number of lower case charecters is",upperCount(name))
    elif c==6:
        name=input("enter file name with extension")
        print("number of special charecters is",lineCount(name))
    else:
        i=1
        break
```
